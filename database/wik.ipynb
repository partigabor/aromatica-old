```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send an HTTP request to the Wiktionary webpage
url = 'https://en.wiktionary.org/wiki/cardamom'
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
# Find the translations section
translations_section = soup.find('span', {'id': 'Translations'})

# Extract the translations into a pandas dataframe
translations = []
for sibling in translations_section.find_next_siblings():
    if sibling.name == 'h2':
        break
    if sibling.name == 'h3':
        language = sibling.get_text()
    if sibling.name == 'ul':
        for li in sibling.find_all('li'):
            translations.append({'Language': language, 'Translation': li.get_text()})

translations_df = pd.DataFrame(translations)
print(translations_df)
```
